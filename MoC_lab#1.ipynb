**1. P(C) (6)**

```python
def ciphertext_probability(C):
    C_probability = 0
    pairs_of_indexes = []
    
    for prob_M in range(len(prob_06.iloc[0])):
        for prob_K in range(len(prob_06.iloc[1])):
            if table_06.iloc[prob_M, prob_K] == C: # C = 0...19
                pairs = []
                pairs.append(prob_K)
                pairs.append(prob_M)
                pairs_of_indexes.append(pairs)
    
    for i in range(len(pairs_of_indexes)):
        C_probability += prob_06.iloc[0, pairs_of_indexes[i][0]] * prob_06.iloc[1, pairs_of_indexes[i][1]]
    
    return round(C_probability, 4)


C_prob = []
for C in range(20):
    C_prob.append(ciphertext_probability(C))
```
