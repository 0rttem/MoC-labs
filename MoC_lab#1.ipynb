**1. P(C) (6)**

```python
def ciphertext_probability(C):
    C_probability = 0
    pairs_of_indexes = []
    
    for prob_M in range(len(prob_06.iloc[0])):
        for prob_K in range(len(prob_06.iloc[1])):
            if table_06.iloc[prob_M, prob_K] == C: # C = 0...19
                pairs = []
                pairs.append(prob_K)
                pairs.append(prob_M)
                pairs_of_indexes.append(pairs)
    
    for i in range(len(pairs_of_indexes)):
        C_probability += prob_06.iloc[0, pairs_of_indexes[i][0]] * prob_06.iloc[1, pairs_of_indexes[i][1]]
    
    return round(C_probability, 4)


C_prob = []
for C in range(20):
    C_prob.append(ciphertext_probability(C))
```

```python
C_distribution = pd.DataFrame(C_prob).T
C_distribution
```

```python
check_sum = 0
for i in range(len(C_distribution.iloc[0])):
    check_sum += C_distribution.iloc[0, i]

print(check_sum)
```

**2. P(M,C) (6)**

```python
def joint_probability_MC(C, t):
    res = []
    
    for j in range(len(table_06.iloc[0])):
        if table_06.iloc[j, t] == C:
            res.append(prob_06.iloc[1, j] * prob_06.iloc[0, t])

    return sum(res)

MC_prob = []
r = []
for t in range(20):
    for C in range(20):
        MC_prob.append(joint_probability_MC(C, t))
    
        r.append(MC_prob)
```
